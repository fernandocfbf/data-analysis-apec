## Pergunta (s) a ser respondida

Pergunta: O fato dos países se juntarem a APEC faz com que o PIB cresça e diminua consequentemente o índice de probreza?

## Explicação da hipótese e mecanismo

APEC é uma organização que tem por objetivo proporcionar o crescimento econômico dos países da região Ásia-Pacífico, através de uma área de livre comércio entre os países membros. Desde modo é esperado que os indicadores de economia e os derivados do mesmo, dos países membros, apresentem melhores resultados ao passar dos anos.

```python
#importando as bibliotecas necessárias
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#importanto o excel

poverty = pd.read_excel('poverty.xls')

```

```python
poverty1= poverty.set_index('Country')
```

```python
#código para selecionar apenas os anos necessários, ou seja, momentos depois do Chile entrar no bloco.

poverty_foc = poverty1.loc[['Chile','Mexico','Peru','Thailand','Canada','Philippines','Indonesia','Australia']]
poverty_columns = poverty_foc[poverty_foc.columns[37:58]]
poverty_tranpose = poverty_columns.transpose()
poverty_tranpose
```

```python
# calculando a média anual dos países no bloco

media = pd.Series(poverty_tranpose.transpose().mean())
media
```

```python
#Criando lista do Chile para plotar um gráfico de Poverty x Tempo
Chile = pd.Series(poverty_tranpose.Chile)



```

```python
#Criando uma lista de tempo para ser usada no gráfico
tempo = np.arange(1997,2018)

#plotando a média anual dos países pelos dados obtidos do Chile
plt.plot(tempo, Chile,'red', label="Chile")
plt.plot(tempo, media, 'blue', label="Média anua dos países da APEC")
plt.xlabel("tempo")
plt.ylabel("People in poverty (%)")
plt.legend()
plt.grid(True)
plt.show()
```
