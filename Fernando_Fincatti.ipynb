## Primeira etapa

Nessa primeira etapa do projeto, analisaremos a média do índice de pobreza dos países da APEC e compararemos com a média de um grupo controle no tempo. O período de anos escolhido foi de 1997 até 2017 pois foi o intervalo de tempo em que haviam menos dados faltando na base de dados.

```python
#importando as bibliotecas necessárias
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
import os
print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
#importanto o excel

poverty = pd.read_excel('poverty.xls', sheet_name="Data", skiprows=3)

```

```python
#transformando a tabela

poverty1= poverty.set_index('Country')
```

```python
#vamos selecionar os países da APEC e uma série temporal de 1997 até 2017.
#fazendo isso, poderemos calcular uma média entre os países da APEC durantes essa série temporal.

APEC = ['Canada', 'Chile', 'China', 'United States', 'Philippines', 'Indonesia', 'Malaysia', 'Mexico', 'Peru', 'Russian Federation', 'Thailand', 'Vietnam']
poverty1.loc[:,"APEC"] = "não"
```

```python
#colocando sim em todos os países que são da APEC

poverty1.loc[APEC, "APEC"] = "sim"
```

```python
#Vamos criar a tabela filtradas do grupo APEC de 1997 até 2017

grupo_APEC = poverty1.loc[poverty1.APEC == 'sim']
grupo_APEC_foc = grupo_APEC[grupo_APEC.columns[37:58]]
grupo_APEC_foc_transpose = grupo_APEC_foc.transpose()
grupo_APEC_serie = grupo_APEC_foc_transpose
grupo_APEC_serie_tranpose = grupo_APEC_serie.transpose()
```

```python
#calculando a média anual dos países no bloco

media_poverty_APEC = pd.Series(grupo_APEC_serie.transpose().mean())
```

Agora que temos uma média anual dos países da APEC, precisamos de um grupo controle, ou seja, países fora do bloco que possam servir de comparação. O critério de seleção para estes países é primeiramente regional, ou seja, países que compartilham a mesma região do grupo selecionado da APEC. Em caso de não haver nenhum país próximo que não pertença ao grupo, será ultilizado uma tabela dos rankings das economias mundiais, e será escolhido países comparáveis.

```python
#vamos selecionar outros 6 países, e criar sua tabela

fora_APEC = ["Ecuador","Egypt, Arab Rep.","Germany","Netherlands","Belgium","Pakistan"]
fora_APEC1 = poverty1.loc[fora_APEC]
fora_APEC_foc = fora_APEC1[fora_APEC1.columns[37:58]]
fora_APEC_foc_transpose = fora_APEC_foc.transpose()
fora_APEC_serie = fora_APEC_foc_transpose
fora_APEC_serie_transpose = fora_APEC_serie.transpose()
```

```python
#Criando lista do Chile para plotar um gráfico de Poverty x Tempo

media_poverty_fora_APEC = pd.Series(fora_APEC_serie.transpose().mean())
```

```python
#Criando uma lista de tempo para ser usada no gráfico

tempo = np.arange(1997,2018)

#plotando a média anual dos países da APEC e do grupo de controle no tempo

plt.plot(tempo, media_poverty_APEC,'red', label="Média anual dos países da APEC")
plt.plot(tempo, media_poverty_fora_APEC, 'blue', label="Média anual dos países fora da APEC")
plt.xlabel("tempo")
plt.ylabel("People in poverty (%)")
plt.legend()
plt.grid(True)
plt.show()
```

Com o gráfico acima podemos ver claramente uma diminuição quase linear do índice de pobreza nos países da APEC, enquanto no grupo controle houveram períodos de oscilações mais constantes, ou seja, membros da APEC apresentam diminuição constante no índice de pobreza.

## Segunda etapa

Nessa segunda etapa do projeto, analisaremos a média do crescimento em porcentagem do PIB per capita dos países da APEC e compararemos com a média do mesmo grupo controle no tempo.

```python
#Vamos criar a tabela filtrada do grupo APEC de 1997 até 2017

grupo_APEC_gdp = gdp1.loc[gdp1.APEC == 'sim']
grupo_APEC_foc_gdp = grupo_APEC_gdp[grupo_APEC.columns[37:58]]
grupo_APEC_foc_transpose_gdp = grupo_APEC_foc_gdp.transpose()
grupo_APEC_serie_gdp = grupo_APEC_foc_transpose_gdp
grupo_APEC_serie_tranpose_gdp = grupo_APEC_serie_gdp.transpose()
```

Com base na análise dos dados acima, podemos ver que de forma geral os países membros da APEC mantém um índice de crescimento mais alto quando comparado ao grupo de controle. 
Entretando, temos no ano de 1998 uma queda muito brusca na média da APEC. Este fato pode ser explicado pela crise asiática que afetou alguns países, entre eles a Indonesia, que por sua vez teve um índice de crescimento baixíssimo (-14%).
Outro período que representa uma queda brusca no crescimento é o ano de 2009. Este acontecimento pode ser explicado através da crise mundial de 2008-2009, onde a maior economia do mundo, os Estados Unidos, sofreu uma instabilidade e refletiu em grande parte dos outros países no globo.

## Terceira etapa

Nessa terceira etapa do projeto serão avaliados simultaneamente os índices de pobreza e crescimento PIB per capita. Para isso foram escolhidos dois anos, e os resultados obtidos serão comparados. 

```python
#Vamos fazer uma tabela especial para GDP

gdp_2000 = gdp1.loc[:, '2000'].to_frame()

```

```python
#Vamos fazer uma tabela especial para poverty

poverty_2000 = poverty1.loc[:, '2000'].to_frame()
```

```python
#Joint nas tabelas

data = gdp_2000.join(poverty_2000, how="inner", lsuffix ="_GDP", rsuffix='_poverty')
```

```python
# Adicionando a coluna para ver se pertence ou não ao bloco

data.loc[:,"APEC"] = 'não'
data.loc[APEC, "APEC"] = 'sim'
```

```python
plt.scatter(data.loc[data.APEC=="sim","2000_GDP"], data.loc[data.APEC=="sim", "2000_poverty"], color="r", label="países da APEC")
plt.scatter(data.loc[data.APEC=="não","2000_GDP"], data.loc[data.APEC=="não", "2000_poverty"], color="b", label="resto do mundo")
plt.title('GDP X poverty')
plt.legend()
plt.xlabel('PIB per capita growth %')
plt.ylabel('People in poverty (%)')
plt.grid(True)
```

```python
def reta(dados_x, dados_y, c):    
    a = dados_y.cov(dados_x) / dados_x.var()
    b = dados_y.mean() - a*dados_x.mean()

    print('Ajuste linear:')
    print('Coeficiente linear:', b)
    print('Coeficiente angular:', a)
    plt.scatter(dados_x, dados_y, c=c, alpha=0.8)
    plt.title('GDP X poverty')
    plt.xlabel('PIB per capita growth %')
    plt.ylabel('People in poverty (%)')

    plt.plot((dados_x.min(), dados_x.max()), (a*dados_x.min()+b, a*dados_x.max()+b), color=c)

    plt.tight_layout()
```

```python
plt.scatter(data.loc[data.APEC=="sim","2000_GDP"], data.loc[data.APEC=="sim", "2000_poverty"], color="r")
reta(data.loc[data.APEC=="sim","2000_GDP"], data.loc[data.APEC=="sim", "2000_poverty"], "r",)
plt.scatter(data.loc[data.APEC=="não","2000_GDP"], data.loc[data.APEC=="não","2000_poverty"], color="b")
reta(data.loc[data.APEC=="não","2000_GDP"], data.loc[data.APEC=="não","2000_poverty"], "b",)
plt.title("GDP X poverty")
plt.grid(True)
```

Com o gráfico acima, podemos perceber que os países membros da APEC tendem a ficar na região inferior direita do gráfico, ou seja, apresentam menor índice de pobreza e maior crescimento economico em relação ao países que não pertencem ao grupo, que por sua vez estão na dispersos pelo gráfico. 
Entretanto como os dados obtidos estão muito dispersos, não podemos dizer claramente o que consiste o coeficiente daa reta calculado.

## Conclusão

Analisando os indicadores no tempo, podemos dizer que o crescimento anual do PIB per capita dos países da APEC, em média, é mais significante em comparação aos países do grupo de controle. Assim podemos dizer o mesmo a respeito da pobreza, uma vez que a longo prazo a média de pessoas em situação de pobreza nos países membros da organização é menor do que a média do grupo de controle.

A partir do gráfico de dispersão, temos uma melhor visualização de como é a distrubuição dos países tanto da APEC como do resto do mundo. Apesar de os países da APEC apresentarem crescimento econômico e queda do índice de pobreza simultaneamente, alguns países do resto do mundo não apresentam relação entre queda do índice de pobreza e melhorias nos indicadores econômicos.

Por fim temos a análise do boxplot, a qual reafirma o que foi supracitado. Apesar de os países apresentarem medianas próximas quanto ao crescimento econômico, o índice de pobreza não apresenta melhorias devido a esse fato, ou seja, para alguns países não há relação entre o aumento do PIB per capita e melhoras no índice de pobreza.

Destarte podemos afirmar que o indicador de crescimento de PIB per capita não é o único fator determinante para diminuição da pobreza. Isso pode ser explicado por um divisão desigual do PIB, ou seja, países onde há pessoas que concentram o capital em suas mãos enquanto uma maior parcela da população ainda possui baixa renda.
